```python
import facenet_mini as fm
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow import keras
from sklearn.manifold import TSNE

%matplotlib inline
```

```python
dataset = fm.Dataset("./notebooks/images/train/",n_faces=300)
```

```python
dataset.parse()
```

```python
dataset.x.shape
```

```python
model = keras.models.model_from_json(open("./ckp/facenet.json","r").read())
```

```python
model.load_weights("./ckp/facenet")
```

```python
emb = model.predict(dataset.x)
```

```python
tsne = TSNE(2,)
```

```python
enc = tsne.fit_transform(emb)
```

```python
plt.figure(figsize=(12,12))
sns.scatterplot(enc[:,0],enc[:,1],hue=dataset.y)
plt.legend("off")
```
